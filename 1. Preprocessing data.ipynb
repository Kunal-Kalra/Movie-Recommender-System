```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('movies.csv')
credits = pd.read_csv('credits.csv')
```

```python
movies.head()
```

```python
credits.head()
```

```python
movies = movies.merge(credits, on='title')
```

```python
movies.head()
```

```python
rating = movies[['movie_id', 'vote_average', 'vote_count']]
movies = movies[['movie_id', 'title', 'genres', 'overview', 'keywords', 'cast', 'crew']]
```

```python
rating.head()
```

```python
movies.head()
```

```python
from ast import literal_eval 
#literal_eval is used to covert given string to list for easy exctation
def extract(x):
    List=[]
    for i in literal_eval(x):
        List.append(i['name'])
    return List
```

```python
movies['genres'] = movies['genres'].apply(extract)
movies['keywords'] = movies['keywords'].apply(extract)
```

```python
movies[['genres', 'keywords']].head()
```

```python
def extract_4(x):
    List=[]
    count = 0
    for i in literal_eval(x):
        List.append(i['name'])
        count+=1
        if count >= 4:
            break
    return List

def extract_direc(x):
    List=[]
    for i in literal_eval(x):
        if i['job'] == "Director":
            List.append(i['name'])
            break
    return List
```

```python
movies['cast'] = movies['cast'].apply(extract_4)
movies['crew'] = movies['crew'].apply(extract_direc)
```

```python
movies[['cast', 'crew']].head()
```

```python
movies.head()
```

```python
movie_dataset = movies.merge(rating, on='movie_id')
movie_dataset.head()
```

```python
movie_dataset.to_csv('movies_dataset.csv')
```

```python
movies_dataset.dropna(inplace=True)
```

```python
movies['title'] = movies['title'].apply(lambda x:x.split())
movies['overview'] = movies['overview'].apply(lambda x:x.split()) 
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x]) 
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x]) 
```
